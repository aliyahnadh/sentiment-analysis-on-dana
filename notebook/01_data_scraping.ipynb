Pada tahap ini dilakukan proses pengumpulan data berupa ulasan pengguna aplikasi DANA yang diperoleh dari platform Google Play Store. Pengambilan data dilakukan secara mandiri menggunakan bahasa pemrograman Python dengan memanfaatkan library `google-play-scraper`, yang memungkinkan ekstraksi ulasan aplikasi secara langsung tanpa menggunakan API resmi.

```python
from google_play_scraper import Sort, reviews
import pandas as pd
```

## Proses Scraping
Proses scraping dilakukan dengan mengambil 15.000 ulasan terbaru dari aplikasi DANA.

```python
print("Sedang mengambil 15000 ulasan...")

result, continuation_token = reviews(
    'id.dana',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=15000
)
```

```python
# Lihat hasil
df = pd.DataFrame(result)
print(f"\nTotal ulasan: {len(df)}")
```

```python
df.info()
```

```python
df.head()
```

## Seleksi Fitur Data
Dari seluruh atribut yang diperoleh, hanya beberapa kolom yang relevan untuk analisis sentimen yang dipilih, yaitu:

*   `reviewId`: ID unik ulasan
*   `content`: isi teks ulasan
*   `score`: rating yang diberikan pengguna (1–5)





```python
df_dana = df[['reviewId', 'content', 'score']]
df_dana.head()
```

## Simpan Dataset
Dataset hasil scraping kemudian disimpan dalam format `.csv` agar dapat digunakan pada tahap selanjutnya, seperti preprocessing dan pelatihan model.

```python
df_dana.to_csv("dana_reviews.csv", index=False)
```
